## 3. customerData
product_rating against variables in customerData

```python
# merging
pc = customerData.merge(po[["product_id", "customer_id", "product_rating"]], on="customer_id")
pc.drop_duplicates(subset=["product_id", "customer_id"], inplace=True) # we do not want duplicates where the same customer buys multiple items of the same product
pc.info()
```

```python
sb.catplot(data = pc, x="product_rating", y="customer_state")

fig = plt.figure(figsize = (14,10))
sb.boxplot(x = "product_rating", y= "customer_state", data = pc)
```

## 4. sellerData
product_rating against variables in sellerData

```python
# merging
ps = sellerData.merge(po[["product_id", "seller_id", "product_rating"]], on="seller_id")
ps.drop_duplicates(subset=["product_id"], inplace=True) # we do not want duplicates of the same products 
ps.info()
```

```python
sb.catplot(data = ps, x="product_rating", y="seller_state")

fig = plt.figure(figsize = (14,10))
sb.boxplot(x = "product_rating", y= "seller_state", data = ps)
```
